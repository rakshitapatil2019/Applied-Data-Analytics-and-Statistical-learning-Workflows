```python
data["Total Marks"] = (data["Assignment_Score"] + data["Internal_Marks"] + data["Final_Exam_Marks"])
data
```

```python
data["External Marks"] = (data["Assignment_Score"] + data["Internal_Marks"]) - data["Final_Exam_Marks"]
data
```

```python
data["total_Percentage"] = (data["Total_Marks"]/ 100)*100
data
```

```python
def grade(Total_Marks):
  if Total_Marks >= 90:
    return "A"
  elif Total_Marks >= 80:
    return "B"
  elif Total_Marks >= 70:
    return "C"
  elif Total_Marks >= 60:
    return "D"
  elif Total_Marks >= 50:
    return "E"
  else:
    return "F"
data["Grade"] = data["Percentage"].apply(grade)
data
```
